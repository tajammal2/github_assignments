# Body Mass Index (BMI)
- weight
- height

```python
weight=float(input("What is your weight in Kgs? "))
# weight=int(weight)
height=float(input("What is your height in meters? "))
# height=float(height)
BMI=weight/height**2
name=input("What is your name? ")
print("my name is:", name, "and my BMI is:", BMI)
```

### Cat Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)

titanic = sns.load_dataset("titanic")
sns.catplot(x="sex",y="survived", hue="class", kind="bar", data=titanic)
plt.show()
```

### Countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
titanic=sns.load_dataset("titanic")
p1 = sns.countplot(x="sex", data=titanic, hue="class")
# p1 = sns.countplot(x="sex", data=titanic, hue="class")
p1.set_title("Plot for counting")
plt.show()
```

### Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
titanic=sns.load_dataset("titanic")
g=sns.FacetGrid(titanic, row="sex", hue="alone")
g=(g.map(plt.scatter, "age", "fare").add_legend())
plt.show()
```

```python
# Cat Plot
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
titanic=sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
plt.show()
```

```python
# count plot
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)

titanic=sns.load_dataset("titanic")
p1=sns.countplot(x="sex", hue="class", data=titanic)
p1.set_title("Counting Plot")
plt.show()
```
